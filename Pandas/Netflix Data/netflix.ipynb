## چارت سهم Movie و TV Show

```python
dftype = df['type'].value_counts()
dftype
```

```python
plt.pie(dftype,labels = ['Movie','TV Show'],explode=[0.1,0],autopct='%1.1f%%')
plt.show()
```

## نمودار ده کارگردان اول با بیشترین میزان تولید

```python
dfdirector=df['director'].value_counts().reset_index().sort_values(by='count',ascending = False).head(10)
dfdirector
```

```python
x = dfdirector['director'].loc[1:10]
y = dfdirector['count'].loc[1:10]
```

```python
plt.figure(figsize = (20, 5))
plt.bar(x,y, color ='maroon')
plt.show()
```

## نمودار ده کشور اول با بیشترین میزان تولید

```python
country = df['country'].value_counts().reset_index().sort_values(by='count',ascending = False).head(60)
country
```

```python
a = country['country'].head(10)
b = country['count'].head(10)
```

```python
plt.figure(figsize = (20, 5))
plt.bar(a,b, color ='darkgreen')
plt.show()
```
